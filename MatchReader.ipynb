# Data Collection
League of Legends is a game created by the company known as 'Riot Games'. Riot Games has developed an API that is available through the Riot Games Developer Portal. This portal can be accessed by anyone with a Riot account. The API has methods that request information about different aspects of their game such as the people who play it and the games that are played.

We used a package called RiotWatcher, which has implementation for all public methods available in the Riot API, specificially for python.

League of Legends has a ranked system which places players into the 'appropriate' rank/division based on wins/losses and game performance.
The aim is to use the specialised methods to gather multiple databases of game information which will represent the different ranks.

```python
#get the ranked entries for the bottom of each division in each tier
divisions=['I','II','III','IV']
tiers=['IRON','BRONZE','SILVER','GOLD','PLATINUM','DIAMOND']
entries=[]
for tier in tiers:
    for division in divisions:
        entries.append(lol_watcher.league.entries(my_region,'RANKED_SOLO_5x5',tier,division))
```

```python
count = 0
for tier in tiers:
    for division in divisions:
        print(tier + " " + division +": " + str(len(entries[count])) + " entries")
        count += 1
```

There are **6 overall ranks** ('IRON', 'BRONZE', 'SILVER', 'GOLD', 'PLATINUM', 'DIAMOND'). In each rank there are **4 divisions** which even further splits up the players.
Using the **.league.entries** method we are given the **bottom 205 players for each rank/division**.

```python
#limiting the number of entries per division from 205 to 12 and combining each rank's entries (48 entries per rank)
iron = []
bronze = []
silver = []
gold = []
platinum = []
diamond = []

#count is keeping track of rank/division, counter is keeping track of number of entries per rank
counter = 0
for count, entry in enumerate(entries):
    for summoner in entry:
        if(count < 4 and counter < limit):
            iron.append(summoner)
            counter += 1
            if(counter == limit):
                counter = 0
                break
        elif(count < 8 and counter < limit):
            bronze.append(summoner)
            counter += 1
            if(counter == limit):
                counter = 0
                break
        elif(count < 12 and counter < limit):
            silver.append(summoner)
            counter += 1
            if(counter == limit):
                counter = 0
                break
        elif(count < 16 and counter < limit):
            gold.append(summoner)
            counter += 1
            if(counter == limit):
                counter = 0
                break
        elif(count < 20 and counter < limit):
            platinum.append(summoner)
            counter += 1
            if(counter == limit):
                counter = 0
                break
        elif(count < 24 and counter < limit):
            diamond.append(summoner)
            counter += 1
            if(counter == limit):
                counter = 0
                break
        
rank = {
    "iron": iron,
    "bronze": bronze,
    "silver": silver,
    "gold": gold,
    "platinum": platinum,
    "diamond": diamond
}
```

Since each database represents a rank, we need to combine the divisions within each rank. The above code performs this task while also **dropping the number of entries included from 205 to 12** due to the excessive size of the databases that would come from 820 (4*205) players per rank.

```python
#getting puuid from player entries
for r in rank:
    for count, player in enumerate(rank[r]):
        counter = count%48
        try:
            rank[r][counter] = lol_watcher.summoner.by_name(my_region,rank[r][counter]['summonerName'])
        except:
            print(rank[r].pop(counter))
            continue
    print("FINISHED: " + r)
```

In order to get the match history of each player we have to get the player's **puuid** which can be found by using the **.summoner.by_name** method on each of the entries.

```python
#number of players in each rank
for r in rank:
    print(r + " players: " + str(len(rank[r])))
```

```python
#getting the last 20 matchIds for each player using their puuid
iron = []
bronze = []
silver = []
gold = []
platinum = []
diamond = []

matches = {
    "iron": iron,
    "bronze": bronze,
    "silver": silver,
    "gold": gold,
    "platinum": platinum,
    "diamond": diamond
}

for r in rank:
    for count, player in enumerate(rank[r]):
        counter = count%48
        try:
            temp=lol_watcher.match_v5.matchlist_by_puuid(shard,rank[r][counter]['puuid'])
            for match in temp:
                matches[r].append(match)
        except:
            continue
```

Using a method called **.match_v5.matchlist_by_puuid** we get the last 20 games of the player whose **puuid** we have specified.
The 20 games are in the form of IDs which are attached to a dictionary that we have initialised.

```python
#fetching the match information from matchIds and filtering for ranked (has very long runtime)
iron = []
bronze = []
silver = []
gold = []
platinum = []
diamond = []

ranked = {
    "iron": iron,
    "bronze": bronze,
    "silver": silver,
    "gold": gold,
    "platinum": platinum,
    "diamond": diamond
}

for r in matches:
    for match in matches[r]:
        try:
            game = lol_watcher.match_v5.by_id(shard,match)
            if(game['info']['queueId'] == 420):
                ranked[r].append(game)
        except:
            continue
    print("FINISHED " + r)
```

The **.match_v5.by_id** gets the game information using each games ID number. The format of the output has been demonstrated earlier in the code. We also filter the games by whether or not they are ranked (**queueId == 420**) since there are other game modes in the game.

```python
#number of ranked matches appended to each rank
for r in ranked:
    print(r + ": " + str(len(ranked[r])))
```

Since not every game played in a player's last 20 games are ranked, there is inconsistency in database sizes between ranks.

It is observed that, higher the rank, greater the proportion of games from their last 20 matches are ranked (iron 334, diamond 492). This may imply that iron players care less about ranked and enjoy/focus on other aspects of the game more.

## Generating the Databases
We are creating a csv file for each rank. The fields will be found through filtering the match information (nested dictionary) for what we need.

```python
#creating new csv file for each rank
for r in ranked:
    filename = 'Data/'+ r + '.csv'
    with open(filename, 'w', newline = '') as file:
        writer = csv.writer(file)
        writer.writerow(['matchId','gameDuration','visionScore','wardsBought','consumablesPurchased','damageDealtToChampions',
                         'totalHealing','timeSpentDead','ccTimeDealt','minionKills','totalGold','championLevel','summonerLevel',
                         'spell1Casts','spell2Casts','spell3Casts','spell4Casts','summoner1Casts','summoner2Casts',
                         'firstBaron','baronKills','firstDragon','dragonKills','firstInhibitor','inhibitorKills',
                         'firstTower','towerKills','damageToObjectives','firstBlood','championKills',
                         'firstRiftHerald','riftHeraldKills',
                         'champion1','champion2','champion3','champion4','champion5',
                         'enemyChampion1','enemyChampion2','enemyChampion3','enemyChampion4','enemyChampion5',
                         '1ban1','1ban2','1ban3','1ban4','1ban5','2ban1','2ban2','2ban3','2ban4','2ban5', 'teamId', 'win'])
```

```python
#each game shows information for the 10 players that participate (5 on winning side, 5 on losing side)
#filtering game information to create a record for each team 
for r in ranked:
    filename = 'Data/'+ r + '.csv'
    with open(filename, 'a', newline = '') as file:
        for match in ranked[r]:
            
            #duration is inconsistently written in seconds or milliseconds 
            duration = match['info']['gameDuration']
            if(len(str(duration)) > 5):
                duration = int(duration/1000)
            
            #initialising variables that need to be accumulated from each team's participants
            winVisionScore = 0
            winWardsBought = 0
            winConsumablesBought = 0
            winDamageDealtToChampions = 0
            winHealing = 0
            winTimeDead = 0
            winCCTimeDealt = 0
            winMinionKills = 0
            winTotalGold = 0
            winChampionLevel = 0
            winSummonerLevel = 0
            winSpell1Cast = 0
            winSpell2Cast = 0
            winSpell3Cast = 0
            winSpell4Cast = 0
            winSummoner1Casts = 0
            winSummoner2Casts = 0
            winDamageToObjectives = 0
            winChamps = []
            
            loseVisionScore = 0
            loseWardsBought = 0
            loseConsumablesBought = 0
            loseDamageDealtToChampions = 0
            loseHealing = 0
            loseTimeDead = 0
            loseCCTimeDealt = 0
            loseMinionKills = 0
            loseTotalGold = 0
            loseChampionLevel = 0
            loseSummonerLevel = 0
            loseSpell1Cast = 0
            loseSpell2Cast = 0
            loseSpell3Cast = 0
            loseSpell4Cast = 0
            loseSummoner1Casts = 0
            loseSummoner2Casts = 0
            loseDamageToObjectives = 0
            loseChamps = []
         
            #accumulating winning and losing team stats
            for participant in match['info']['participants']:
                if(participant['win'] == True):
                    winVisionScore += participant['visionScore']
                    winWardsBought += participant['visionWardsBoughtInGame']
                    winConsumablesBought += participant['consumablesPurchased']
                    winDamageDealtToChampions += participant['totalDamageDealtToChampions']
                    winHealing += participant['totalHeal']
                    winTimeDead += participant['totalTimeSpentDead']
                    winCCTimeDealt += participant['totalTimeCCDealt']
                    winMinionKills += participant['totalMinionsKilled']
                    winTotalGold += participant['goldEarned']
                    winChampionLevel += participant['champLevel']
                    winSummonerLevel += participant['summonerLevel']
                    winSpell1Cast += participant['spell1Casts']
                    winSpell2Cast += participant['spell2Casts']
                    winSpell3Cast += participant['spell3Casts']
                    winSpell4Cast += participant['spell4Casts']
                    winSummoner1Casts += participant['summoner1Casts']
                    winSummoner2Casts += participant['summoner2Casts']
                    winDamageToObjectives += participant['damageDealtToObjectives']
                    winChamps.append(participant['championId'])


                else:
                    loseVisionScore += participant['visionScore']
                    loseWardsBought += participant['visionWardsBoughtInGame']
                    loseConsumablesBought += participant['consumablesPurchased']
                    loseDamageDealtToChampions += participant['totalDamageDealtToChampions']
                    loseHealing += participant['totalHeal']
                    loseTimeDead += participant['totalTimeSpentDead']
                    loseCCTimeDealt += participant['totalTimeCCDealt']
                    loseMinionKills += participant['totalMinionsKilled']
                    loseTotalGold += participant['goldEarned']
                    loseChampionLevel += participant['champLevel']
                    loseSummonerLevel += participant['summonerLevel']
                    loseSpell1Cast += participant['spell1Casts']
                    loseSpell2Cast += participant['spell2Casts']
                    loseSpell3Cast += participant['spell3Casts']
                    loseSpell4Cast += participant['spell4Casts']
                    loseSummoner1Casts += participant['summoner1Casts']
                    loseSummoner2Casts += participant['summoner2Casts']
                    loseDamageToObjectives += participant['damageDealtToObjectives']
                    loseChamps.append(participant['championId'])

                    
            for team in match['info']['teams']:
                if(team['win'] == True):
                    try:
                        winTeamId = team['teamId']
                        winFirstBaron = team['objectives']['baron']['first']
                        winBaronKills = team['objectives']['baron']['kills']        
                        winFirstDragon = team['objectives']['dragon']['first']
                        winDragonKills = team['objectives']['dragon']['kills']
                        winFirstInhibitor = team['objectives']['inhibitor']['first']
                        winInhibitorKills = team['objectives']['inhibitor']['kills']
                        winFirstTower = team['objectives']['tower']['first']
                        winTowerKills = team['objectives']['tower']['kills']
                        winFirstBlood = team['objectives']['champion']['first']
                        winChampionKills = team['objectives']['champion']['kills']
                        winFirstRiftHerald = team['objectives']['riftHerald']['first']
                        winRiftHeraldKills = team['objectives']['riftHerald']['kills']
                        win = team['win']
                        winBan1 = team['bans'][0]['championId']
                        winBan2 = team['bans'][1]['championId']
                        winBan3 = team['bans'][2]['championId']
                        winBan4 = team['bans'][3]['championId']
                        winBan5 = team['bans'][4]['championId']
                    except:
                        continue

                        
                else:
                    try:
                        loseTeamId = team['teamId']
                        loseFirstBaron = team['objectives']['baron']['first']
                        loseBaronKills = team['objectives']['baron']['kills']        
                        loseFirstDragon = team['objectives']['dragon']['first']
                        loseDragonKills = team['objectives']['dragon']['kills']
                        loseFirstInhibitor = team['objectives']['inhibitor']['first']
                        loseInhibitorKills = team['objectives']['inhibitor']['kills']
                        loseFirstTower = team['objectives']['tower']['first']
                        loseTowerKills = team['objectives']['tower']['kills']
                        loseFirstBlood = team['objectives']['champion']['first']
                        loseChampionKills = team['objectives']['champion']['kills']
                        loseFirstRiftHerald = team['objectives']['riftHerald']['first']
                        loseRiftHeraldKills = team['objectives']['riftHerald']['kills']
                        lose = team['win']
                        loseBan1 = team['bans'][0]['championId']
                        loseBan2 = team['bans'][1]['championId']
                        loseBan3 = team['bans'][2]['championId']
                        loseBan4 = team['bans'][3]['championId']
                        loseBan5 = team['bans'][4]['championId']
                    except:
                        continue


            writer = csv.writer(file)
            
            #winning side record
            writer.writerow([match['metadata']['matchId'],
                            duration,
                            winVisionScore,
                            winWardsBought,
                            winConsumablesBought,
                            winDamageDealtToChampions,
                            winHealing,
                            winTimeDead,
                            winCCTimeDealt,
                            winMinionKills,
                            winTotalGold,
                            winChampionLevel,
                            winSummonerLevel,
                            winSpell1Cast, winSpell2Cast, winSpell3Cast, winSpell4Cast, 
                            winSummoner1Casts, winSummoner2Casts,
                            winFirstBaron, winBaronKills, 
                            winFirstDragon, winDragonKills,
                            winFirstInhibitor, winInhibitorKills,
                            winFirstTower, winTowerKills,
                            winDamageToObjectives,
                            winFirstBlood, winChampionKills,
                            winFirstRiftHerald, winRiftHeraldKills,
                            winChamps[0],winChamps[1],winChamps[2],winChamps[3],winChamps[4],
                            loseChamps[0],loseChamps[1],loseChamps[2],loseChamps[3],loseChamps[4],
                            winBan1,winBan2,winBan3,winBan4,winBan5,
                            loseBan1,loseBan2, loseBan3,loseBan4,loseBan5, winTeamId, win])
 
            #losing side record
            writer.writerow([match['metadata']['matchId'],
                            duration,
                            loseVisionScore,
                            loseWardsBought,
                            loseConsumablesBought,
                            loseDamageDealtToChampions,
                            loseHealing,
                            loseTimeDead,
                            loseCCTimeDealt,
                            loseMinionKills,
                            loseTotalGold,
                            loseChampionLevel,
                            loseSummonerLevel,
                            loseSpell1Cast, loseSpell2Cast, loseSpell3Cast, loseSpell4Cast, 
                            loseSummoner1Casts, loseSummoner2Casts,
                            loseFirstBaron, loseBaronKills, 
                            loseFirstDragon, loseDragonKills,
                            loseFirstInhibitor, loseInhibitorKills,
                            loseFirstTower, loseTowerKills,
                            loseDamageToObjectives,
                            loseFirstBlood, loseChampionKills,
                            loseFirstRiftHerald, loseRiftHeraldKills,
                            loseChamps[0],loseChamps[1],loseChamps[2],loseChamps[3],loseChamps[4],
                            winChamps[0],winChamps[1],winChamps[2],winChamps[3],winChamps[4],
                            winBan1,winBan2,winBan3,winBan4,winBan5,
                            loseBan1,loseBan2, loseBan3,loseBan4,loseBan5, loseTeamId, lose])
```

Since the game information shows information about the performance of all 10 individual players in the game, we have created two records per match which has the accumulated values for each field for each team. We felt the need to do this so our models could have a target (win = true or false).

## Functions for Reading Libraries 
The libraries read, have values stored that are related to different ID's for fields that are or were in the databases. For example 1ban3 = 99 is the 3rd player from team 1's ban, which in this case was 'Lux'.

```python
import json

def printLibrary(dic,dict_keys,list_keys=[]):
    for key in dic.keys():
        if key in dict_keys:
            print("{0}: {1}".format(key,list(dic.get(key).keys())))
        elif key in list_keys:
            print("{0}: {1}".format(key,dic.get(key)[0]))
        else:
            print("{0}: {1}".format(key,dic.get(key)))
```

```python
with open('Libraries/champion.json', encoding="utf8") as f:
    champion = json.load(f)

printLibrary(champion,['data'])
```

```python
with open('Libraries/item.json', encoding="utf8") as f:
    item = json.load(f)

printLibrary(item,['basic','data'],['groups','tree'])
```
